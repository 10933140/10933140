```python
import plotly
import pandas as pd
# Load data
#df = pd.read_csv("finance-charts-apple.csv")
df = pd.read_csv("https://raw.githubusercontent.com/tewei0328/trade-backtest/main/0-data/2330.csv")
```

```python
df.head()
```

```python
import plotly.graph_objects as go
line1 = go.Scatter(x=df['date'], y=df['Close'], name='Close')
fig  = go.Figure(line1)
fig.show()
plotly.offline.plot(fig, filename='lifeExp.html')
```

```python
df['date'] = pd.to_datetime(df['date'])
df['date'] > pd.to_datetime('2012-01-01')
```

```python
df1=df[df['date'] > pd.to_datetime('2012-01-01')]
```

```python
line1 = go.Scatter(x=df1['date'], y=df1['Close'], name='Close')
fig  = go.Figure(line1)
fig.show()
plotly.offline.plot(fig, filename='lifeExp1.html')
```
